# **WORD FREQUENCY COUNTER**

**Practice Prjoect**

```python
#WORD FREQUENCY COUNTER

#text to work on
text='I am learning practical skills in AI so that I can be an AI specialist who build tools for a cause'

#tokenize the text
#using split function and storing the tokenized word in a var
words = text.split()

#store how many times each word appears
word_count={} #use an empty dict

#loop thru every word and count it
for word in words:
  if word in word_count:
    word_count[word] = word_count[word] + 1 #if word alr exists in dict, add 1 to its count
  else:
    word_count[word] = 1 #if the word is new, start its count at 1

print(word_count) #run
```

```python
#find most used word

#goes thru every word in dict and finds the one with highest count.
most_common = max(word_count, key = word_count.get) #max compare by the values(the counts), not the words themselves

print("most used word:", most_common)
print("count:", word_count[most_common])
```

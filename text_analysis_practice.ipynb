```python
text = "AI is changing the world. AI helps doctors. AI helps teachers. The world is becoming smarter. Learning AI is important."
```

```python
#1 Count how many times each word appears

words = text.split()
word_count={}

for word in words:
  if word in word_count:
    word_count[word]=word_count[word] +1
  else:
    word_count[word]=1
print(word_count)
```

```python
#2 Prints only words that appear more than once

for word in word_count:
  if word_count[word] > 1:
    print(word, "->", word_count[word])
```

```python
#3 Tells the total number of unique words

print("unique words:",len(word_count))
```
